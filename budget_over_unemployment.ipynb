## Basic Data Visualization using high level percentages (not using entire CSV)

```python
columns = ['true_exps', 'just_for_fun','biz_exp','qual_of_life']

pre_move = [35.3,11.9,9,13.4]
biz_is_booming = [36.5,19.1,13,13.5]
unemployed = [54,11.7,22.7,12.7]
back_to_werk = [34.4, 10.5,38.5,8.7]

data = [pre_move, biz_is_booming, unemployed, back_to_werk]

df = pd.DataFrame(data, columns = columns)
df
```

```python
df['work_state'] = ''

df['work_state'] = ['pre_move', 'biz_is_booming', 'unemployed', 'back_to_werk']

new_order = ['work_state','true_exps', 'just_for_fun','biz_exp','qual_of_life']

df = df[new_order]

df.set_index('work_state', inplace=True)
df
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

# Left plot (Bar chart)
df.plot(kind='bar', ax=ax1, width=0.8)
ax1.set_title('Comparison of % Income Spent')
ax1.set_xlabel('Work State')
ax1.set_ylabel('% of Income Spent')
ax1.tick_params(axis='x', rotation=45)
ax1.legend(title='Metrics')

# Right plot (Line chart)
df['true_exps'].plot(ax=ax2, kind='line', marker='o', label='True Expenses')
df['just_for_fun'].plot(ax=ax2, kind='line', marker='^', label='Just for Fun')
df['biz_exp'].plot(ax=ax2, kind='line', marker='s', label='Business Expenses')
df['qual_of_life'].plot(ax=ax2, kind='line', marker='d', label='Quality of Life')
ax2.set_title('Comparison of Metrics Over Work States')
ax2.set_xlabel('Work State')
ax2.set_ylabel('Values')
ax2.legend(title='Metrics')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
df['true_exps'].plot(ax=ax1, kind='line', title='True Expenses Over Time', marker='o')
ax1.set_xlabel('Work State')
ax1.set_ylabel('% of Income Spent')
ax1.tick_params(axis='x', rotation=45)

df['just_for_fun'].plot(ax=ax2, kind='line', title='Just for Fun Over Time', marker='o')
ax2.set_xlabel('Work State')
ax2.set_ylabel('% of Income Spent')
ax2.tick_params(axis='x', rotation=45)

df['biz_exp'].plot(ax=ax3, kind='line', title='Business Expenses Over Time', marker='o')
ax3.set_xlabel('Work State')
ax3.set_ylabel('% of Income Spent')
ax3.tick_params(axis='x', rotation=45)

df['qual_of_life'].plot(ax=ax4, kind='line', title='Quality of Life Over Time', marker='o')
ax4.set_xlabel('Work State')
ax4.set_ylabel('% of Income Spent')
ax4.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
